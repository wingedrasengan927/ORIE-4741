**Q**: What is the geometric intuition of a subgradient?

**A**: For a convex function $f$, a vector $g$ is a subgradient at a point $x$ if the line (or hyperplane) it defines stays below the entire function graph.

---

**Q**: What is the formal definition of a subgradient?

**A**: A vector $g$ is a subgradient of $f: \mathrm{R}^d \to \mathrm{R}$ at a point $x$ if for all other points $y$:
$$
f(y) \geq f(x) + g^T(y - x)
$$

---

**Q**: What is the formal definition of a subdifferential?

**A**: The subdifferential of a function $f: \mathrm{R}^d \to \mathrm{R}$ at a point $x \in \mathrm{R}^d$ is the set of all subgradients of $f$ at $x$:
$$
\partial f(x) = \left\{ g: f(y) \geq f(x) + g^T(y-x) \, \forall \, y \in \mathrm{R}^d \right\}
$$

---

**Q**: Suppose $f: \mathrm{R}^d \to \mathrm{R}$ is differentiable. What is the subdifferential of $f$ at a point $x$?

**A**:
$$
\partial f(x) = \left\{ \nabla f(x) \right\}
$$

---

**Q**: What is the linearity property of subgradients

**A**: If 
$$h = f + g,$$
then
$$
g_1 + g_2 \in \partial h(x),
$$

where $g_1 \in \partial f(x)$ and $g_2 \in \partial g(x)$

---

**Q**: What is the chain rule property of subdifferentials

**A**: Suppose
$$h = f \circ g,$$

where $g: \mathrm{R}^d \to \mathrm{R}$ is differentiable and $f: \mathrm{R} \to \mathrm{R}$. Then,
$$
\partial h(x) = \partial f\big( g(x) \big) \nabla g(x).
$$

---

**Q**: Let
$$
h = l \circ z,
$$
where $z(w) = w^Tx$ and $l(y, z)$ is a convex loss function. Assuming $y$ is fixed in $l$, Find the partial of $h$ at $w$

**A**: Applying the chain rule for subdifferentials, we get:
$$
\begin{align*}
\partial h(w) &= \partial l\big( z(w) \big) \nabla z(w) \\[4pt]
&= \partial l \left( w^Tx \right) \nabla \left( w^Tx \right) \\[4pt]
&= x \cdot \partial l \left( w^Tx \right)
\end{align*}
$$

### Generative Questions

**Q**: Suppose a function $f: \mathrm{R} \to \mathrm{R}$ is convex. If it's differentiable on an open neighbourhood of a point $x$ but not at $x$, then how would you find the subdifferential of $f$ at $x$

**A**: First we find the left and right derivatives of $f$ at $x$:
$$
\begin{align*}
g^+ &= \lim_{\epsilon \to 0} \nabla f(x + \epsilon) \\[4pt]
g^- &= \lim_{\epsilon \to 0} \nabla f(x - \epsilon)
\end{align*}
$$

Then, the subdifferential of $f$ at $x$ is given by:
$$
\partial f(x) = \left\{ \alpha g^+ + (1 - \alpha) g^-: \alpha \in [0, 1] \right\} 
$$

---

Let's work through some examples

**Q**: Let $f(x) = |x|$. Find the subdifferential of $f$ at $x = 0$

**A**: Using the above method, we get:
$$
\partial f(0) = [ -1, 1 ]
$$

---

**Q**: Let $l(y, z) = (y - z)^2$. Assuming $y$ is fixed, find the subdifferential of $l$

**A**: $l$ is differentiable everywhere, so at any point, the subdifferential of $l$ is simply the gradient of $l$ at that point 

---

**Q**: Let $l(y, z) = |y - z|$. Assuming $y$ is fixed, find the subdifferential of $l$

**A**: $l$ is differentiable everywhere except at $z=y$. So let's find the subdifferential of $l$ at $y$. Using the above method, we get:
$$
\partial l(y) = [ y - 1, y + 1 ]
$$

---

**Q**: Let $l(y, z) = (1 - yz)_{+}$. Assuming $y$ is fixed, find the subdifferential of $l$

**A**: $l$ is differentiable everywhere except at $yz=1$. So let's find the subdifferential of $l$ at $1/y$. Let's break it down:

case $y < 0$:
In this case we have $g^+ = -y$ and $g^- = 0$. Thus the partial of $l$ is given by
$$
\partial l \left( \frac{1}{y} \right) = [0, -y]
$$

case $y > 0$:
In this case we have $g^+ = 0$ and $g^- = -y$. Thus the partial of $l$ is given by
$$
\partial l \left( \frac{1}{y} \right) = [-y, 0]
$$

---

**Q**: Let $l(y, z) = \log\big(1 + \exp(-yz) \big)$. Assuming $y$ is fixed, find the subdifferential of $l$

**A**: $l$ is differentiable everywhere, so at any point, the subdifferential of $l$ is simply the gradient of $l$ at that point

### Proximal Gradient Method

**Q**: What kind of functions can be minimized with the proximal gradient method?

**A**: Composite functions of the form
$$
f(\theta) + g(\theta),
$$

where $f$ is differentiable and $g$ is not

---

Consider the function
$$
F(\theta) = f(\theta) + g(\theta),
$$

where $f$ is differentiable and $g$ is not.

**Q**: Rewrite $F$ by replacing $f$ with its linear approximation at iteration $k$

**A**:
$$
\begin{align*}
F(\theta) &\approx f(\theta_k) + \nabla f(\theta_k)^T (\theta - \theta_k) + g(\theta) \\[4pt]
&= \hat{F} (\theta; \theta_k)
\end{align*}
$$

---

**Q**: At the next iteration, we want to pick $\theta$ such that the above expression is minimized. Furthermore, we want to ensure that $\theta$ is close to $\theta^k$. What is the penalty term that we need to add to the above expression to ensure that $\theta$ is close to $\theta^k$?

**A**:
$$
\frac{1}{2h_k} \cdot \| \theta - \theta_k \|_2^2,
$$

where $h_k > 0$ is the step size

---

Our goal is to minimize the objective
$$
\hat{F} (\theta; \theta_k) + \frac{1}{2h_k} \cdot \| \theta - \theta_k \|_2^2
$$

The objective can be expanded and simplified as follows:
$$
\begin{align*}
\hat{F} (\theta; \theta_k) + \frac{1}{2h_k} \cdot \| \theta - \theta_k \|_2^2 &= \\[10pt]
f(\theta_k) + \nabla f(\theta_k)^T (\theta - \theta_k) + g(\theta) + \frac{1}{2h_k} \cdot \| \theta - \theta_k \|_2^2 &= \\[10pt]
f(\theta_k) + \nabla f(\theta_k)^T \theta - f(\theta_k)^T \theta_k + g(\theta) + \frac{1}{2h_k} \cdot \left( \| \theta \|^2 - 2\theta_k^T \theta + \| \theta_k \|^2 \right)
\end{align*}
$$

Removing the constant terms, the expression simplifies to:
$$
\begin{align*}
g(\theta) + \nabla f(\theta_k)^T \theta + \frac{1}{2h_k} \cdot \left( \| \theta \|^2 - 2\theta_k^T \theta \right) &= \\[10pt]
g(\theta)  + \frac{1}{2h_k} \cdot \left(  \| \theta \|^2 - 2\theta_k^T \theta + 2h_k \nabla f(\theta_k)^T \theta \right) &= \\[10pt]
g(\theta)  + \frac{1}{2h_k} \cdot \left(  \| \theta \|^2 - 2 \left(\theta_k  -h_k \nabla f(\theta_k) \right)^T \theta \right) &= \\[10pt]
g(\theta)  + \frac{1}{2h_k} \cdot \left(  \| \theta \|^2 - 2 \left(\theta_k  -h_k \nabla f(\theta_k) \right)^T \theta + \| \theta_k  -h_k \nabla f(\theta_k) \|^2 \right) \\[4pt] - \frac{1}{2h_k} \| \theta_k  -h_k \nabla f(\theta_k) \|^2 \quad &= \\[10pt]
g(\theta)  + \frac{1}{2h_k} \cdot \| \theta - \left( \theta_k  -h_k \nabla f(\theta_k) \right) \|^2  - \frac{1}{2h_k} \| \theta_k  -h_k \nabla f(\theta_k) \|^2 \\[10pt]
\end{align*}
$$

Removing the constant term, the objective simplifies to:
$$
g(\theta)  + \frac{1}{2h_k} \cdot \| \theta - \left( \theta_k  -h_k \nabla f(\theta_k) \right) \|^2
$$

Let
$$
\theta_{k + \frac{1}{2}} = \theta_k  -h_k \nabla f(\theta_k)
$$

Now our objective can be expressed as:
$$
g(\theta)  + \frac{1}{2h_k} \cdot \| \theta - \theta_{k + 1/2}  \|^2
$$

$\theta_{k + 1/2}$ is called an intermediate iterate. It is the result of applying the gradient update rule exclusively to the differentiable component, $f(θ)$.
