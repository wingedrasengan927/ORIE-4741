**Q**: What is the general form of the regularised empirical risk?

**A**:
$$
\sum_{i=1}^N l\big( y_i, \hat{f}(x_i) \big) + r(w),
$$
where
- $r$ is the regularizer
- $w$ is the parameter vector of $\hat{f}$

---

**Q**: What is a disadvantage of LASSO regression when compared to Ridge regression?

**A**: LASSO regression is relatively unstable

---

**Q:** Purely in terms of coefficient magnitude, why is Ridge regression more stable than LASSO regression?

**A:** Because Ridge uses an $L_2$ penalty, which penalizes large coefficients more strongly

**Q:** How does LASSO regression handle correlated features?

**A:** LASSO tends to select one correlated feature and shrink the others to zero.

---

**Q:** How does Ridge regression handle correlated features?

**A:** Ridge tends to spread the weight across correlated features.

---

**Q**: What is a disadvantage of Ridge regression when compared to LASSO regression?

**A**: Ridge regression splits weights evenly between correlated features, while LASSO regression selects only one of the correlated features

**Q**: What is the general form of a non-negative regularizer?

**A**: 
$$r(w) = \sum_{i=1}^d \mathbf{1}(w_i \geq 0),$$

where $\mathbf{1}$ is the convex indicator defined as:
$$
\mathbf{1}(z)=
\begin{cases}
0, & \text{$z$ is true},\\
\infty, & \text{$z$ is false}.
\end{cases}
$$

---

**Q**: What is the general form of the objective function of a non-negative least squares problem?

**A**: 
$$
\sum_{i=1}^N (y_i - w^Tx_i)^2 + \sum_{j=1}^d \mathbf{1}(w_j \geq 0),
$$

where $\mathbf{1}$ is the convex indicator defined as:
$$
\mathbf{1}(z)=
\begin{cases}
0, & \text{$z$ is true},\\
\infty, & \text{$z$ is false}.
\end{cases}
$$

---

**Q**: How would you implement the non-negative regularizer in NumPy?

**A**: See the function `non_neg_regularizer` below

---

```python
import numpy as np

def non_neg_regularizer(w: np.ndarray) -> float:
    return 0.0 if np.all(w >= 0) else np.inf
```

**Q**: What is the general form of a smooth regularizer?

**A**: 
$$\sum_{j=1}^{d-1} (w_{j+1} - w_j)^2,$$

where $w \in \mathrm{R}^d$

---

**Q**: How can you express the smooth regularizer in matrix form?

**A**: 
$$\| Dw \|^2,$$

where $w \in \mathrm{R}^d$ and $D \in \mathrm{R}^{(d-1) \times d}$. The entries of the matrix $D$ are defined as follows:
$$
D_{ij} = \begin{cases}
-1, \quad &\text{if } j = i, \\[4pt]
1, \quad &\text{if } j = i+1, \\[4pt]
0, \quad &\text{otherwise}.
\end{cases}
$$

---

**Q**: How would you implement the smooth regularizer in NumPy?

**A**: See the function `smooth_regularizer` below

---

```python
def smooth_regularizer(w: np.ndarray) -> float:
    w_backward = w[:-1]
    w_forward = w[1:]
    return np.linalg.norm(w_forward - w_backward)**2
```

**Q**: What is the gradient of the smooth regularizer?

**A**:
$$\nabla r(w) = 2D^T Dw,$$
where $D \in \mathrm{R}^{(d-1)\times d}$ is the difference matrix.

**Explanation**: Since

$$r(w)=|Dw|^2 = w^T D^T D w,$$

and $D^T D$ is symmetric, its gradient is
$$\nabla r(w)=2D^\top Dw.$$

### Generative Questions

**Q**: Show that least squares regression is scale invariant

**A**: Suppose $X \in \mathrm{R}^{N \times d}$ is the data matrix and $y \in \mathrm{R}^N$ is target vector. Let $X^\prime = XD$, where $D \in \mathrm{R}^{d \times d}$ is a diagonal matrix with non-zero diagonal entries. In other words, $X^\prime$ is the scaled version of $X$. If we assume that the columns of $X$ are linearly independent, then the solution to the least squares problem:
$$
\text{minimize} \quad \| y -Xw \|^2
$$
is given by:
$$
\hat{w} = (X^TX)^{-1}X^Ty
$$
The columns of $X^\prime$ will also be linearly independent. Thus, the solution to the least squares problem:
$$
\text{minimize} \quad \| y -X^\prime w^\prime \|^2
$$
is given by:
$$
\hat{w}^\prime = \big( (X^\prime)^T X^\prime \big)^{-1}(X^\prime)^Ty
$$
The corresponding predictions are given by:
$$
\begin{align*}
X^\prime \hat{w}^\prime &= X^\prime \big( (X^\prime)^T X^\prime \big)^{-1}(X^\prime)^Ty \\[4pt]
&= X^\prime \big( (XD)^T XD \big)^{-1}(XD)^Ty \\[4pt]
&= X^\prime \big( D^TX^T XD \big)^{-1}(D^TX^T)y \\[4pt]
&= X^\prime D^{-1} \big( X^TX \big)^{-1}(D^T)^{-1}D^TX^Ty \\[4pt]
&= X^\prime D^{-1} \big( X^TX \big)^{-1}X^Ty \\[4pt]
&= X^\prime D^{-1} \hat{w} \\[4pt]
&= XDD^{-1} \hat{w} \\[4pt]
&= X \hat{w} \\[4pt]
\end{align*}
$$
Thus both give the same predictions

Q.E.D

**Q**: Show that ridge regression is not scale invariant

**A**: Suppose $X \in \mathrm{R}^{N \times d}$ is the data matrix and $y \in \mathrm{R}^N$ is target vector. Let $X^\prime = XD$, where $D \in \mathrm{R}^{d \times d}$ is a diagonal matrix with non-zero diagonal entries. In other words, $X^\prime$ is the scaled version of $X$. Here we don't need to assume the columns of $X$ are linearly independent. The solution to the regularized least squares problem:
$$
\text{minimize} \quad \| y -Xw \|^2 + \lambda \| w \|^2
$$
is given by:
$$
\hat{w} = (X^TX + \lambda I)^{-1}X^Ty
$$
The columns of $X^\prime$ also don't need to be linearly independent. The solution to the regularized least squares problem:
$$
\text{minimize} \quad \| y -X^\prime w^\prime \|^2 + \lambda \| w^\prime \|^2
$$
is given by:
$$
\hat{w}^\prime = \big( (X^\prime)^T X^\prime + \lambda I \big)^{-1}(X^\prime)^Ty
$$
The corresponding predictions are given by:
$$
\begin{align*}
X^\prime \hat{w}^\prime &= X^\prime \big( (X^\prime)^T X^\prime + \lambda I \big)^{-1}(X^\prime)^Ty \\[4pt]
&= X^\prime \big( (XD)^T XD + \lambda I \big)^{-1}(XD)^Ty \\[4pt]
&= X^\prime \big( D^TX^T XD + \lambda I \big)^{-1}(D^TX^T)y
\end{align*}
$$

The above expression cannot be simplified further. Thus the predictions aren't the same

Q.E.D

**Q:** What does it mean when we say ordinary linear regression is scale invariant?

**A:** Scaling the features does not change the predictions. The coefficients adjust to compensate for the scaling.

---

**Q**: What does it mean when we say Ridge regression is not scale invariant?

**A:** Scaling the features can change the predictions. The coefficients cannot fully adjust for the scaling because of the $L_2$ penalty.

---

Thus, to get the same answer no matter the units of measurement, it's important to standardize the data

**Q**: Why should features usually be standardized before Ridge regression?

**A**: Because Ridge is not scale invariant.
