## Cards

Q: Suppose $w$ minimizes $\| y - Xw \|^2$. Then, what are the normal equations that $w$ must satisfy?

A: $$X^TXw = X^Ty$$

---

Q: How can you express the objective function
$$
\| y - Xw \|^2 + \lambda \| w \|^2
$$
where $\lambda > 0$ is the regularisation parameter as
$$
\| y^\prime - X^\prime w \|^2.
$$

A: 
$$
\left\|
\begin{bmatrix}
y \\
0
\end{bmatrix} - 
\begin{bmatrix}
X \\
\sqrt{\lambda} I
\end{bmatrix}
w
\right\|^2
$$

---

Q: Show that the matrix
$$
X^TX + \lambda I
$$
is invertible for any $X$ and $\lambda > 0$

A: The matrix can be decomposed as:
$$
\begin{align*}
X^TX + \lambda I &=
\begin{bmatrix}
X^T & \sqrt{\lambda} I
\end{bmatrix}
\begin{bmatrix}
X \\
\sqrt{\lambda} I
\end{bmatrix} \\[4pt]
&= \begin{bmatrix}
X \\
\sqrt{\lambda} I
\end{bmatrix}^T
\begin{bmatrix}
X \\
\sqrt{\lambda} I
\end{bmatrix}
\end{align*}
$$
Now it suffices to show that the matrix
$$
\begin{bmatrix}
X \\
\sqrt{\lambda} I
\end{bmatrix}
$$
is invertible, which is clearly the case

#### Generative Questions

Q: Suppose that $X \in \mathrm{R}^{n \times d}$ and $\operatorname{Rank} \left( X \right) = r \leq d$. If the SVD of $X$ is given by:
$$
X = U \Sigma V^T,
$$
then how can the normal equations be rewritten?

A: We have:
$$
\begin{align*}
X^TXw &= X^Ty \\[4pt]
\implies \left( U \Sigma V^T \right)^T \left( U \Sigma V^T \right) w &= \left( U \Sigma V^T \right)^Ty \\[4pt]
\implies \left( V \Sigma U^T \right) \left( U \Sigma V^T \right) w &= \left( V \Sigma U^T \right)y \\[4pt]
\implies \left( V \Sigma^2 V^T \right) w &= \left( V \Sigma U^T \right)y \\[4pt]
\implies \left( \Sigma^{-2} V^T \right) \left( V \Sigma^2 V^T \right) w &= \left( \Sigma^{-2} V^T \right) \left( V \Sigma U^T \right)y \\[4pt]
\implies V^T w &= \Sigma^{-1} U^T y
\end{align*}
$$

Since $V^T$ isn't invertible, we cannot solve for $w$ directly

---

Q: The normal equations expressed in terms of the SVD of $X$ is given by:
$$
V^T w = \Sigma^{-1} U^T y
$$
What is a solution and why it isn't unique? Assume that $X \in \mathrm{R}^{n \times d}$ and $\operatorname{Rank} \left( X \right) = r < d$

A: Clearly
$$
w = V \Sigma^{-1}U^Ty
$$
satisfies the normal equations. To show the solution isn't unique, it suffices to show that $\operatorname{null} V^T \neq \{ 0 \}$; because then if $v \in \operatorname{null} V^T$ and $v \neq 0$, then $w + \alpha v$ also satisfies the normal equations for any $\alpha \in \mathrm{R}$

Since $\operatorname{Rank} \left( X  \right) < d$, we have $\operatorname{null} X \neq \{ 0 \}$. This implies $\exists \, v \in \mathrm{R}^d. \, Xv = 0, \, v \neq 0$. Therefore, we have:
$$
\begin{align*}
0 &= Xv \\[4pt]
&= \left( U \Sigma V^T \right) v \\[4pt]
&= \left( \Sigma^{-1} U^T \right) \left( U \Sigma V^T \right) v \\[4pt]
&= V^T v
\end{align*}
$$
Thus $\operatorname{null} V^T \neq \{ 0 \}$

Q.E.D
