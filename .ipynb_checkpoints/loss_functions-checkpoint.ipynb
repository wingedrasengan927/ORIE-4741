## Losses for classification

**Q**: What is the Neyman-Pearson metric?

**A**: The Neyman-Pearson metric is simply a weighed average of false positive rate and false negative rate

---

**Q**: What is the general form of the loss function for boolean classification

**A**: If $y = 1$, then we want to penalize negative $\hat{y}$, and if $y = -1$, then we want to penalize positive $\hat{y}$

---

**Q**: What is the formula for the squared loss for boolean classification?

**A**: 
$$
l(y, \hat{y}) = 
\begin{cases}
(1 + \hat{y})^2, \quad &\text{if } y=-1, \\[4pt]
\kappa (1 - \hat{y})^2, \quad &\text{if } y=1.
\end{cases}
$$

where $\kappa > 0$ is our relative dislike for false negatives

---

**Q**: What is the problem with the squared loss for boolean classification?

**A**: The squared loss penalizes $\hat{y}$ if it's too far from $+1$ or $-1$ even when it has the correct sign

---

**Q**: What is the formula for the Neyman-Pearson loss for boolean classification?

**A**: 
$$
l(y, \hat{y}) = 
\begin{cases}
\mathrm{1} (\hat{y} \geq 0), \quad &\text{if } y=-1, \\[4pt]
\kappa \cdot \mathrm{1} (\hat{y} \leq 0), \quad &\text{if } y=1.
\end{cases}
$$

where $\kappa > 0$ is our relative dislike for false negatives and $\mathrm{1}$ is an indicator function defined as follows:
$$
\mathrm{1} (z) = 
\begin{cases}
1, \quad &\text{if } z= \text{true}, \\[4pt]
0, \quad &\text{if } z= \text{false}.
\end{cases}
$$

---

**Q**: What is the problem with the Neyman-Pearson loss for boolean classification?

**A**: It's hard to minimize as the function is discontinuous and has a derivative of zero almost everywhere

---

**Q**: What is the formula for the sigmoid loss for boolean classification?

**A**: 
$$
l(y, \hat{y}) = 
\begin{cases}
1 \div \left( 1 + e^{-\hat{y}} \right), \quad &\text{if } y=-1, \\[6pt]
\kappa \div \left( 1 + e^{\hat{y}} \right), \quad &\text{if } y=1.
\end{cases}
$$

where $\kappa > 0$ is our relative dislike for false negatives.

---

**Q**: What is the formula for the logistic loss for boolean classification?

**A**: 
$$
l(y, \hat{y}) = 
\begin{cases}
\log \left( 1 + e^{\hat{y}} \right), \quad &\text{if } y=-1, \\[4pt]
\kappa \log \left( 1 + e^{-\hat{y}} \right),\quad &\text{if } y=1.
\end{cases}
$$

where $\kappa > 0$ is our relative dislike for false negatives.

---

**Q**: What is the formula for the hinge loss for boolean classification?

**A**: 
$$
l(y, \hat{y}) = 
\begin{cases}
\left( 1 + \hat{y} \right)_+, \quad &\text{if } y=-1, \\[4pt]
\kappa\left( 1 - \hat{y} \right)_+,\quad &\text{if } y=1.
\end{cases}
$$

where $\kappa > 0$ is our relative dislike for false negatives.

---

**Q**: What is the key characteristic of the hinge loss for boolean classification?

**A**: It adds a margin to the decision boundary

---

**Q**: What is a differentiable but non-convex approximation of the Neyman-Pearson loss for boolean classification?

**A**: Sigmoid loss

---

**Q**: What is a differentiable and convex approximation of the Neyman-Pearson loss for boolean classification?

**A**: Logistic loss

---

**Q**: What is a differentiable approximation of the hinge loss for boolean classification?

**A**: Hubristic loss

---

**Q**: Suppose $a, b \in \mathrm{R}^m$. Find the hyperplane that exactly separates $a$ and $b$

**A**: Let $\hat{y}$ be a point on the hyperplane that exactly separates $a$ and $b$. Then,
$$
\begin{align*}
\| \hat{y} - b \| = \| \hat{y} - a \| \\[4pt]
\implies \| \hat{y} - b \|^2 - \| \hat{y} - a \|^2 = 0 \\[4pt]
\implies \| b \|^2 - 2b^T \hat{y} - \| a \|^2 + 2a^T \hat{y} = 0 \\[4pt]
\implies 2(b - a)^T \hat{y} = \| b \|^2 - \| a \|^2 \\[4pt]
\end{align*}
$$
The above equation represents a hyperplane with normal vector $(b-a)$ passing through the point $(a + b)/2$

Let's verify the hyperplane passes through the point $(a + b)/2$. Substituting
$$
\hat{y} = \frac{a + b}{2},
$$
we get:
$$
\begin{align*}
2(b - a)^T \hat{y} &= (b - a)^T(b + a) \\[4pt]
&= (b^T - a^T)(b + a) \\[4pt]
&= \| b \|^2 + b^Ta - a^Tb - \| a \|^2 \\[4pt]
&= \| b \|^2 - \| a \|^2
\end{align*}
$$

Hence verified

---

**Q**: Suppose $a, b \in \mathrm{R}^m$. How can you check whether a point $\hat{y}$ is closer to $a$ than $b$

**A**: By checking if
$$
2(b - a)^T \hat{y} < \| b \|^2 - \| a \|^2
$$

---

**Q**: Suppose $a, b \in \mathrm{R}^m$. How is the signed distance of a point $\hat{y} \in \mathrm{R}^m$ to the hyperplane that exactly separates $a$ and $b$ defined?

**A**:
$$
D(\hat{y}, a, b) = \frac{2(b - a)^T \hat{y} - \| b \|^2 + \| a \|^2}{2 \| b - a \|}
$$

---

**Q**: Suppose $\psi_1, \ldots, \psi_M \in \mathrm{R}^m$. How can you check which of these vectors is closest to a point $\hat{y} \in \mathrm{R}^m$?

**A**: $\hat{y}$ is closest to $\psi_i$ for some $i \in \{ 1, \ldots, M \}$ if
$$
D(\hat{y}, \psi_i, \psi_j) < 0
$$
for each $j = 1, \ldots, M$ and $j \neq i$. 

Here, $D(\hat{y}, \psi_i, \psi_j)$ represents the signed distance of $\hat{y}$ to the hyperplane that exactly separates $\psi_i$ and $\psi_j$

---

**Q**: Suppose $\psi_1, \ldots, \psi_M \in \mathrm{R}^m$ are classes. What is the general structure of a loss function for multi-class classification?

**A**: 
$$
l(\hat{y}, \psi_i) = \begin{cases}
0, &\, \text{if } \underset{j \neq i}{\max} D_{ij}(\hat{y}) < 0, \\[4pt]
1, &\, \text{otherwise}.
\end{cases}
$$

Here, $D(\hat{y}, \psi_i, \psi_j)$ represents the signed distance of $\hat{y}$ to the hyperplane that exactly separates $\psi_i$ and $\psi_j$

---

**Q**: What is the formula for the hinge loss for multi-class classification?

**A**: 
$$
l(\hat{y}, \psi_i) = \underset{j \neq i}{\max} \left[ \left( 1 + D_{ij}(\hat{y}) \right)_+ \right]
$$

Here, $D(\hat{y}, \psi_i, \psi_j)$ represents the signed distance of $\hat{y}$ to the hyperplane that exactly separates $\psi_i$ and $\psi_j$

---

**Q**: What is the formula for the logistic loss for multi-class classification?

**A**: 
$$
l(\hat{y}, \psi_i) = \log \left( \sum_{j=1}^K \exp \left( D_{ij}(\hat{y}) \right) \right)
$$

Here, $D(\hat{y}, \psi_i, \psi_j)$ represents the signed distance of $\hat{y}$ to the hyperplane that exactly separates $\psi_i$ and $\psi_j$

---

**Q**: How is the log-sum-exp function defined?

**A**: The function $f: \mathrm{R}^n \to \mathrm{R}$ defined by:
$$
f(x) = \log \left( \sum_{j=1}^n \exp(x_j) \right)
$$
is called the log-sum-exp function

---

**Q**: What is a convex and differentiable approximation to the $\max$ function?

**A**: The log-sum-exp function

---

**Q**: What is a probabilistic classifier?

**A**: A probabilistic classifier produces a probabilistic distribution of the output classes for a given input

---

**Q**: Given a dataset
$$
\{ (u_1, v_1), (u_2, v_2), \ldots, (u_n, v_n) \},
$$
how is the likelihood of a probabilistic classifier $G$ defined?

**A**:
$$
\prod_{i=1}^n \hat{p}_i (v_i),
$$

where $\hat{p}_j = G(u_j)$ for each $j$

---

**Q**: Given a dataset
$$
\{ (u_1, v_1), (u_2, v_2), \ldots, (u_n, v_n) \},
$$
how is the negative log likelihood of a probabilistic classifier $G$ defined?

**A**:
$$
-\sum_{i=1}^n \log\hat{p}_i (v_i),
$$

where $\hat{p}_j = G(u_j)$ for each $j$

---

**Q**: Given a dataset
$$
\{ (u_1, v_1), (u_2, v_2), \ldots, (u_n, v_n) \},
$$
how is the optimal constant probabilistic classifier defined?

**A**: Define
$$
q(v) = \frac{1}{n} \times \text{number of data points with $v_j = v$}
$$

for each $v \in \mathcal{V}$

---

**Q**: How is the cross entropy loss of a probabilistic classifier $\hat{p}$ defined?

**A**:
$$
l^{ce}(\hat{p}, v) = -\log \hat{p}(v)
$$

---

**Q**: What is the formula for the logistic loss for multi-class probabilistic classification?

**A**: 
$$
l(\hat{y}, \psi_i) = \log \left( \sum_{j=1}^K \exp \left( D_{ij}(\hat{y}) \right) \right)
$$

Here, $D(\hat{y}, \psi_i, \psi_j)$ represents the signed distance of $\hat{y}$ to the hyperplane that exactly separates $\psi_i$ and $\psi_j$

---

**Q**: What is the relationship between logistic loss and cross entropy loss for multi-class probabilistic classifiction?

**A**: When cross entropy loss is applied to the output of the softmax function, it takes the form of the multi-class logistic loss

---

**Q**: What is the relationship between logistic loss and cross entropy loss for boolean probabilistic classifiction?

**A**: When cross entropy loss is applied to the output of the sigmoid function, it takes the form of the boolean logistic loss

---

**Q**: Given the output space $\mathcal{Y} = \{ 0, 1, \ldots, 7 \}$, what is the binary encoding of the class $7$?

**A**: $$\left[ 1, 1, 1 \right]$$

---

**Q**: What is the general idea of performing multi-class classification using binary classifiers?

**A**: Encode the target class as a vector and use binary classifiers to predict each entry of the vector

---

**Q**: Suppose you're performing multi-class classification using binary classifiers. How would you compute the loss for a data point?

**A**: Encode the target class of the data point as a vector and sum the individual losses for each entry of the vector

---

**Q**: Suppose the output space $\mathcal{Y}$ has $k$ classes. Then, what is the dimension required to binary encode these classes?

**A**: $$\lceil \log_2(k) \rceil$$

---

**Q**: What is the problem with using the quadratic loss for ordinal regression?

**A**: Suppose the output space has 5 classes $\mathcal{Y} = \{ 0, 1, 2, 3, 4 \}$. If the target class is 4, and our predictor outputs 17, then the quadratic loss heavily penalises this prediction This shouldn't be the case

---

**Q**: How to perform ordinal regression via multi-class classification?

**A**: Encode the target class using bigger-vs-smaller encoding. Then, predict each entry of the encoded vector using binary classifiers

---

**Q**: Suppose the output space has 5 classes $\mathcal{Y} = \{ 0, 1, 2, 3, 4 \}$. What is bigger-vs-smaller encoding of the class $3$?

**A**: $$\left[ 1, 1, 1, -1 \right]$$

---

**Q**: If you want to treat ordinal regression as plain regression, which loss should you ideally use?

**A**: Ordinal hinge loss

---

**Q**: Consider a multi-class classification problem with output space $\mathcal{Y}$. If the output of a (non-probabilistic) classifier is $\hat{z}$, then how would you map $\hat{z}$ to a target class in the output space?

**A**: 
$$
\hat{y} = \underset{y \in \mathcal{Y}}{\operatorname{argmin}} l (y, \hat{z}),
$$

where $l$ is the loss function
