**Q**: How to handle missing values of a feature using a boolean indicator?

**A**: Create a new boolean column for that feature that is `true` if the corresponding value isn't missing and `false` otherwise

**Q**: What is the idea behind a low rank model in unsupervised learning?

**A**: Given an $n \times d$ data table $Y$ with rank $r \leq n, d$, the idea is to express $Y$ as a product of two matrices $X \in \mathrm{R}^{n \times r}$ and $W \in \mathrm{R}^{r \times d}$:
$$
\begin{bmatrix} \, \\ X \\ \, \end{bmatrix} \begin{bmatrix} \, & W & \, \end{bmatrix}
\approx
\begin{bmatrix} \, & \, & \, \\ \, & Y & \, \\ \, & \, & \,  \end{bmatrix}
$$

---

**Q**: What is the idea behind PCA?

**A**: Given an $n \times d$ data table $Y$ with rank $r \leq n, d$, the idea is to find two matrices $X \in \mathrm{R}^{n \times r}$ and $W \in \mathrm{R}^{r \times d}$ such that the following objective is minimized:
$$
\| Y - XW \|_F^2
$$

---

**Q**: What is the mathematical idea behind generalized low rank model?

**A**: Given an $n \times d$ data table $Y$ with rank $r \leq n, d$, the idea is to find two matrices $X \in \mathrm{R}^{n \times r}$ and $W \in \mathrm{R}^{r \times d}$ such that the following objective is minimized:
$$
\sum_{(i, j) \in \Omega} l_j \left( Y_{ij}, x_i^Tw_j \right) + \sum_{i=1}^n r_i (x_i) + \sum_{j=1}^d \tilde{r}_j (w_j),
$$

where

- $\Omega$ is a set that contains the indices of non-missing entries
- $l_j$ is the loss function for column $j$
- $r$ and $\tilde{r}$ are regularizers

---

**Q**: Given an $n \times d$ data table $Y$, and you've found $X \in \mathrm{R}^{n \times r}$ and $W \in \mathrm{R}^{r \times d}$ using GLRM. How would you impute the $(i, j)$-th missing entry of $Y$?

**A**: 
$$
\hat{Y}_{ij} = \underset{a}{\operatorname{argmin}} l_j (x_i^Tw_j, a)
$$

### Generative questions

**Q**: Given $Y \in \mathrm{R}^{n \times d}$, how can you minimize
$$
\| Y - XW \|_F^2
$$
with $X \in \mathrm{R}^{n \times r}$ and $W \in \mathrm{R}^{r \times d}$ using the Eckart-Young-Mirsky theorem

**A**: We have
$$
\operatorname{rank} \left( XW \right) \leq r
$$
The Eckart-Young-Mirsky theorem says the $\operatorname{rank} r$ matrix that minimizes
$$
\| Y - Z \|_F^2
$$
is given by
$$
Z = \sum_{j=1}^r \sigma_j u_j v_j^T,
$$
where
$$
Y = \sum_{j=1}^{\operatorname{rank} Y} \sigma_j u_j v_j^T
$$
is the thin SVD of $Y$. We can represent $Z$ as:
$$
Z = U_r \Sigma_r V_r^T,
$$
and then take:
$$
X = U_r
$$
and 
$$
W = \Sigma_r V_r^T
$$

Q.E.D

**Q**: Given $Y \in \mathrm{R}^{n \times d}$, how can you minimize
$$
\| Y - XW \|_F^2
$$
with $X \in \mathrm{R}^{n \times r}$ and $W \in \mathrm{R}^{r \times d}$ using the alternating minimization technique?

**A**: Initialize $W_0$. Then, for each $t = 1, 2, 3, \ldots$ alternate between
$$
X^t = \underset{X}{\text{argmin}} \| Y - XW^{t-1} \|_F^2,
$$
and
$$
W^t = \underset{W}{\text{argmin}} \| Y - X^tW \|_F^2,
$$
until convergence

**Q**: Given $Y \in \mathrm{R}^{n \times d}$ and $X \in \mathrm{R}^{n \times r}$. Find $W \in \mathrm{R}^{r \times d}$ that minimizes
$$
\| Y - XW \|_F^2
$$

**A**: This is matrix least squares problem and the solution is given by:
$$
W^\prime = X^\dagger Y
$$

**Q**: Given $Y \in \mathrm{R}^{n \times d}$ and $W \in \mathrm{R}^{r \times d}$. Find $X \in \mathrm{R}^{n \times r}$ that minimizes
$$
\| Y - XW \|_F^2
$$

**A**: We have:
$$
\begin{align*}
\| Y - XW \|_F^2 &= \| (Y - XW)^T \|_F^2 \\[4pt]
&= \| Y^T - W^TX^T \|_F^2 \\[4pt]
\end{align*}
$$

This is matrix least squares problem and the solution is given by:
$$
\left( X^T \right)^\prime = \left( W^T \right)^\dagger Y^T
$$

Taking the transpose of both sides, we get:

$$
X^\prime = YW^\dagger
$$
