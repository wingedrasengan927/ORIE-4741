## Convergence of Steepest descent

### Assumptions

- $f: \mathrm{R}^n \to \mathrm{R}$
- $f$ is continuously differentiable
- $\nabla f$ is Lipschitz continuous
- $f$ is bounded below

At iteration $x_k$, the curvature condition for step size $\alpha_k$ is given by:
$$ - \nabla f_{k+1}^T p_{k} \leq - c_2 \cdot \nabla f_k^Tp_k,$$
where $c_2 \in (0, 1)$. Subtracting $\nabla f_k^T p_k$ from both sides and multiplying by $-1$, we get:
$$ \left( \nabla f_{k+1} - \nabla f_{k} \right)^T p_{k} \geq (c_2 - 1) \cdot \nabla f_k^Tp_k$$
Using Cauchy-Schwarz inequality, we get:
$$ \left( \nabla f_{k+1} - \nabla f_{k} \right)^T p_{k} \leq \| \nabla f_{k+1} - \nabla f_{k} \| \| p_k \|$$
Since $\nabla f$ is Lipschitz continuous, there exists a non-negative real number $L$ such that
$$
\begin{align*}
\| \nabla f_{k+1} - \nabla f_{k} \| &\leq L \| x_{k+1} - x_k \| \\[4pt]
&= L \alpha_k \| p_k \|
\end{align*}
$$
Thus we get:
$$
\alpha_k \geq \frac{c_2 - 1}{L} \cdot \frac{\nabla f_k^T p_k}{\| p_k \|^2},
$$

At iteration $x_k$, the Armijo condition for step size $\alpha_k$ is given by:
$$ f(x_k + \alpha_k p_k) \leq f(x_k) + c_1 \alpha_k \nabla f_k^T p_k,$$
which implies
$$ f(x_k) - f(x_k + \alpha_k p_k) \geq  - c_1 \alpha_k \nabla f_k^T p_k.$$
If $\alpha_k$ satisfies both the conditions, we get:
$$
\begin{align*}
f(x_k) - f(x_k + \alpha_k p_k) &\geq  \frac{c_1(1 - c_2)}{L} \frac{(\nabla f_k^T p_k)^2}{\| p_k \|^2} \\[8pt]
&= w \cdot \| \nabla f_k \|^2 \cdot \cos^2 \theta_k
\end{align*}
$$
where $$w = \frac{c_1(1 - c_2)}{L}$$ is a constant

We have
$$\sum_{k=0}^N \left( f(x_k) - f(x_{k+1}) \right) = f(x_0) - f(x_N).$$
Since $f$ has a lower bound, there exists a real number $C$ such that $C \leq f(x)$ for all $x$. This implies
$$\sum_{k=0}^\infty \left( f(x_k) - f(x_{k+1}) \right) \leq f(x_0) - C.$$
The R.H.S is finite, which implies
$$\sum_{k=0}^\infty w \cdot \| \nabla f_k \|^2 \cdot \cos^2 \theta_k$$
is finite as
$$
\begin{align*}
\sum_{k=0}^\infty w \cdot \| \nabla f_k \|^2 \cdot \cos^2 \theta_k &\leq
\sum_{k=0}^\infty \left( f(x_k) - f(x_{k+1}) \right) \\[4pt]
&\leq f(x_0) - C \\[4pt]
&< \infty
\end{align*}
$$

Thus, if we assume that each $p_k$ is a legitimate descent direction, then
$$ \lim_{k \to \infty} \| \nabla f_k \| = 0 $$

## Rate of Convergence

We'll consider an objective function that has a quadratic form:
$$
f(x) = \frac{1}{2}x^TAx - b^Tx + c,
$$
where $A$ is symmetric and positive definite

We'll use steepest descent with exact line search

At iterate $x_k$, the descent direction is given by:
$$
p_k = - \nabla f_k
$$

In exact line search, $\alpha_k$ minimizes
$$
\phi(\alpha) = f(x_k + \alpha p_k)
$$

Because $\alpha_k$ minimizes $\phi(\alpha)$, we have:
$$
\begin{align*}
0 &= \phi^\prime(\alpha_k) \\[4pt]
&= \nabla f(x_k + \alpha_k p_k)^T p_k \\[4pt]
\end{align*}
$$

We have:
$$
\nabla f(x) = Ax - b,
$$
and
$$
p_k = - \nabla f_k
$$
So we have:
$$
\begin{align*}
0 &= p_k^T \left[ A(x_k + \alpha_k p_k) - b \right] \\[4pt]
&= p_k^T \left[ Ax_k + \alpha_k A p_k - b \right] \\[4pt]
&= -\nabla f_k^T \left[ \nabla f_k  - \alpha_k A \nabla f_k \right] \\[4pt]
&= -\nabla f_k^T \nabla f_k  + \alpha_k \nabla f_k^T A \nabla f_k \\[4pt]
\end{align*}
$$
Thus $\alpha_k$ equals
$$
\alpha_k = \frac{\nabla f_k^T \nabla f_k}{\nabla f_k^T A \nabla f_k}
$$
