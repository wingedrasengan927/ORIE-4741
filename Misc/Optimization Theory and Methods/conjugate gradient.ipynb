## Conjugate Gradient Descent

### Setup

The CG method is used to find the solution to a system of linear equations $Ax = b$, where $A$ is a positive definite matrix

Consider the objective function
$$
f(x) = \frac{1}{2} x^T A x - b^T x + c.
$$
If $A$ is positive definite, then the value of $x$ which minimises $f(x)$ is a solution to the system of linear equations $Ax = b$

**Explanation 1**: 
If $A$ is positive definite, then $f$ is convex, which means $x^*$ minimizes $f$ if and only if
$$
\begin{align*}
0 &= \nabla f(x^*) \\[4pt]
&= Ax^* - b
\end{align*}
$$
Thus $x^*$ minimises $f$ if and only if it's a solution to the system of linear equations $Ax^* = b$

**Explanation 2**:
Assume that $A$ is symmetric and $x^*$ is a solution to the linear system of equations $Ax = b$. Then,
$$
Ax^* = b.
$$
We have:
$$
\begin{align*}
f(x) - f(x^*) &= \frac{1}{2} x^T A x - \frac{1}{2} (x^*)^T A x^*  - b^T ( x - x^*) \\[4pt]
&= \frac{1}{2} x^T A x - \frac{1}{2} (x^*)^T A x^*  - (Ax^*)^T ( x - x^*) \\[4pt]
&= \frac{1}{2} x^T A x - \frac{1}{2} (x^*)^T A x^*  - (x^*)^TA^T ( x - x^*) \\[4pt]
&= \frac{1}{2} x^T A x - \frac{1}{2} (x^*)^T A x^*  - (x^*)^TA^T x + (x^*)^TA^T x^* \\[4pt]
&= \frac{1}{2} x^T A x + \frac{1}{2} (x^*)^T A x^*  - \frac{1}{2}(x^*)^TA^T x  - \frac{1}{2}(x^*)^TA^T x \\[4pt]
&= \frac{1}{2} x^T A x + \frac{1}{2} (x^*)^T A x^*  - \frac{1}{2}(x^*)^TA x  - \frac{1}{2}(Ax^*)^T x \\[4pt]
&= \frac{1}{2} x^T A x + \frac{1}{2} (x^*)^T A x^*  - \frac{1}{2}(x^*)^TA x  - \frac{1}{2}x^T(Ax^*) \\[4pt]
&= \frac{1}{2} (x - x^*)^T A (x - x^*)
\end{align*}
$$
If $A$ is positive definite, then the RHS term is always greater than zero for $x \neq x^*$, which implies that $x = x^*$ minimises $f$

### Convergence Analysis

At iteration $k$, the residual $r_k$ is defined as:
$$
r_k = b - Ax_k,
$$
which, if you observe, is same as $-\nabla f_k$. The error term is defined as:
$$
e_k = x_k - x^*
$$
In this case $Ax^* = b$, which implies:
$$
\begin{align*}
Ae_k &= Ax_k - Ax^* \\[4pt]
&= Ax_k - b \\[4pt]
&= - (b - Ax_k) \\[4pt]
&= -r_k
\end{align*}
$$
Thus $r_k = -Ae_k$

In steepest descent with exact line search, we've seen that the step size $\alpha_k$ at iteration $k$ is given by:
$$
\begin{align*}
\alpha_k &= \frac{\nabla f_k^T \nabla f_k}{\nabla f_k^T A \nabla f_k} \\[4pt]
&= \frac{ r_k^T r_k}{ r_k^T A r_k}
\end{align*}
$$

Since $A$ is positive definite, by the real spectral theorem, there exists an orthonormal basis $v_1, \ldots, v_n$ of $\mathrm{R}^n$ consisting of eigenvectors of $A$. Let $\lambda_1, \dots, \lambda_n \in \mathrm{R}$ denote the corresponding eigenvalues.

The vector $e_k$ can be expressed as a linear combination of $v_1, \ldots, v_n$ i.e., there exists $\epsilon_1, \ldots, \epsilon_n \in \mathrm{R}$ such that:

$$
e_k = \epsilon_1 v_1 + \cdots + \epsilon_n v_n,
$$

which implies

$$
\begin{align*}
r_k &= -Ae_k \\[4pt]
&= - \epsilon_1 \lambda_1 v_1 - \cdots - \epsilon_n \lambda_n v_n,
\end{align*}
$$

We have:
$$
\begin{align*}
r_k^Tr_k &= \| r_k \|^2 \\[4pt]
&= \sum_{i=1}^n \epsilon_i^2 \lambda_i^2
\end{align*}
$$
Furthermore, we have:
$$
\begin{align*}
r_k^TAr_k = \sum_{i=1}^n \epsilon_i^2 \lambda_i^3
\end{align*}
$$
This implies:
$$
\alpha_k = \frac{\sum_{i=1}^n \epsilon_i^2 \lambda_i^2}{\sum_{i=1}^n \epsilon_i^2 \lambda_i^3}
$$

We have:
$$
x_{k+1} = x_k + \alpha_k p_k
$$
Subtracting $x^*$ from both sides, we get:
$$
e_{k+1} = e_k + \alpha_k p_k
$$
Substituting the values of $p_k$ and $\alpha_k$ we get:
$$
e_{k+1} = e_k + \frac{\sum_{i=1}^n \epsilon_i^2 \lambda_i^2}{\sum_{i=1}^n \epsilon_i^2 \lambda_i^3} r_k
$$

#### What if $e_k$ is an eigenvector of $A$?

Suppose $e_k = a v_i$ for some $i \in \{1, \ldots, n\}$ and some $a \in \mathrm{R}$. Then:
$$
\begin{align*}
r_k &= -Ae_k \\[4pt]
&= -a \lambda_i v_i,
\end{align*}
$$
which implies:
$$
\begin{align*}
\alpha_k &= \frac{a^2 \lambda_i^2}{a^3 \lambda_i^3} \\[4pt]
&= \frac{1}{\lambda_i}
\end{align*}
$$
Now we have:
$$
\begin{align*}
e_{k+1} &= e_k + \alpha_k r_k \\[4pt]
&= e_k - \frac{1}{\lambda_i} \cdot a \lambda_i v_i \\[4pt]
&= e_k - a v_i \\[4pt]
&= e_k - e_k \\[4pt]
&= 0
\end{align*}
$$
Thus convergence is achieved in a single step

#### What if all eigenvalues of $A$ are equal?

Suppose $\lambda_1 = \cdots = \lambda_n = \lambda$. Then:
$$
\begin{align*}
r_k &= - Ae_k \\[4pt]
&= - \epsilon_1 \lambda v_1 - \cdots - \epsilon_n \lambda v_n \\[4pt]
&= -\lambda \left( \epsilon_1 v_1 + \cdots + \lambda v_n \right) \\[4pt]
&= -\lambda e_k \\[4pt]
\end{align*}
$$
This implies:
$$
\begin{align*}
\alpha_k &=  \frac{ r_k^T r_k}{ r_k^T A r_k} \\[4pt]
&= \frac{r_k^T r_k}{r_k^T (-\lambda Ae_k)} \\[4pt]
&= \frac{r_k^T r_k}{\lambda r_k^T r_k} \\[4pt]
&= \frac{1}{\lambda}
\end{align*}
$$
Now we have:
$$
\begin{align*}
e_{k+1} &= e_k + \alpha_k r_k \\[4pt]
&= e_k + \frac{1}{\lambda} \cdot -\lambda e_k \\[4pt]
&= e_k - e_k \\[4pt]
&= 0
\end{align*}
$$
Thus convergence is achieved in a single step

### General Convergence

Define the energy norm of a vector as follows:
$$
\| x \|_A = \left( x^TAx \right)^{\frac{1}{2}}
$$

We have:
$$
\begin{align*}
f(x_{k}) - f(x^*) &= \frac{1}{2} \cdot (x_k - x^*)^TA(x_k - x^*) \\[4pt]
&= \frac{1}{2} \cdot e_k^TAe_k \\[4pt]
&= \frac{1}{2} \cdot \| e_k \|_A ^2
\end{align*}
$$
Thus $f(x_k)$ is minimized when $\| e_k \|_A^2$ is minimized

Let's do some further analysis. We have:
$$
\begin{align*}
\| e_{k+1} \|_A^2 &= e_{k+1}^TAe_{k+1} \\[4pt]
&= (e_{k} + \alpha_k r_k)^TA(e_{k} + \alpha_k r_k) \\[4pt]
&= e_{k}^TAe_{k} + \alpha_ke_{k}^TAr_{k} + \alpha_kr_{k}^TAe_{k} + \alpha_k^2 r_{k}^TAr_{k} \\[4pt]
&= \| e_{k} \|_A^2 + \alpha_k^2 \| r_k \|_A^2 + \alpha_k (Ar_{k})^Te_{k} + \alpha_kr_{k}^TAe_{k} \\[4pt]
&= \| e_{k} \|_A^2 + \alpha_k^2 \| r_k \|_A^2 + \alpha_k r_{k}^TA^Te_{k} + \alpha_kr_{k}^TAe_{k} \\[4pt]
&= \| e_{k} \|_A^2 + \alpha_k^2 \| r_k \|_A^2 + \alpha_k r_{k}^TAe_{k} + \alpha_kr_{k}^TAe_{k} \\[4pt]
&= \| e_{k} \|_A^2 + \alpha_k^2 \| r_k \|_A^2 + 2\alpha_k r_{k}^TAe_{k} \\[4pt]
&= \| e_{k} \|_A^2 + \left( \frac{ r_k^T r_k}{ r_k^T A r_k} \right)^2 \| r_k \|_A^2 - 2 \left( \frac{ r_k^T r_k}{ r_k^T A r_k} \right) r_{k}^Tr_k \\[4pt]
&= \| e_{k} \|_A^2 + \frac{ \left( r_k^T r_k \right)^2}{ \| r_k \|_A^4 } \| r_k \|_A^2 - 2 \frac{ \left( r_k^T r_k \right)^2 }{ \| r_k \|_A^2 } \\[4pt]
&= \| e_{k} \|_A^2 - \frac{ \left( r_k^T r_k \right)^2 }{ \| r_k \|_A^2 } \\[4pt]
&= \| e_{k} \|_A^2 - \| e_{k} \|_A^2 \frac{ \left( r_k^T r_k \right)^2 }{ \| e_{k} \|_A^2 \| r_k \|_A^2 } \\[4pt]
&= \| e_{k} \|_A^2 \left( 1 - \frac{ \left( r_k^T r_k \right)^2 }{ \| e_{k} \|_A^2 \| r_k \|_A^2 } \right) \\[4pt]
\end{align*}
$$

Substituting
$$
\begin{align*}
r_k^Tr_k = \sum_{j=1}^n \epsilon_j^2 \lambda_j^2
\end{align*}
$$
and
$$
\begin{align*}
r_k^TAr_k =  \sum_{j=1}^n \epsilon_j^2 \lambda_j^3,
\end{align*}
$$
and
$$
\begin{align*}
\| e_k \|_A^2 &=  e_k^TAe_k \\[4pt]
&=  -e_k^Tr_k \\[4pt]
&= (\epsilon_1 v_1 + \cdots + \epsilon_n  v_n)^T( \epsilon_1 \lambda_1 v_1 + \cdots + \epsilon_n \lambda_n v_n) \\[4pt]
&= \sum_{j=1}^n \epsilon_j^2 \lambda_j
\end{align*}
$$
we get:
$$
\| e_{k+1} \|_A^2 = \| e_{k} \|_A^2 \omega^2,
$$
where
$$
\omega^2 = 1 - \frac{\left( \sum_j \epsilon_j^2 \lambda_j^2 \right)^2}{\left( \sum_j \epsilon_j^2 \lambda_j^3 \right)\left( \sum_j \epsilon_j^2 \lambda_j \right)}
$$

An upper bound for $\omega$ is given by:
$$
\omega \leq \frac{\kappa - 1}{\kappa + 1},
$$
Where $\kappa$ is the condition number of $A$. Thus, we get
$$
\frac{\| e_{k+1} \|_A}{ \| e_k \|_A } \leq \frac{\kappa - 1}{\kappa + 1},
$$
which implies the rate of convergence is Q-linear and worsens as the condition number increases
