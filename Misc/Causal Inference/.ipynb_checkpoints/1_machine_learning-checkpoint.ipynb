**Q**: What is a problem with LASSO regression for feature importance

**A**: Suppose there are two positively correlated features. LASSO pushes one feature to zero, and the effect of that feature is picked up by the other feature. This increases its weight making it look more important

---

**Q**: What is a simple way to get feature importance without relying on model coefficients?

**A**: Sort the model predictions and partition them into $K$ bins of equal sizes. For each bin, compute the average of every feature within that bin. Then, rank the features based on how much their average varies across bins
