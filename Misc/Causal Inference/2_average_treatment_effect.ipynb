**Q**: How is the individual treatment effect defined for the data point $(X_i, W_i, Y_i)$?

**A**: 
$$Y_i(1) - Y_i(0),$$

where $Y_i(1)$ represents the individuals outcome if they had received treatment, and $Y_i(0)$ represents the outcome if they had not received any treatment

---

**Q**: Why is it not possible to calculate the individual treatment effect for the data point $(X_i, W_i, Y_i)$?

**A**: In reality, we either observe $Y_i(1)$ or $Y_i(0)$ but not both

---

**Q**: How is the average treatment effect defined?

**A**: $$\tau = \mathrm{E} \left[ Y_i(1) - Y_i(0) \right]$$

---

**Q**: What is the key characteristic of a randomized control trial?

**A**: The probability that an individual is assigned a treatment arm is fixed and does not depend on their potential outcomes

---

**Q**: How is the propensity score for an individual intuitively defined?

**A**: It is the probability that the individual receives treatment

---

**Q**: How is the propensity score for the data point $(X_i, W_i, Y_i)$ mathematically defined?

$$
e(X_i) = \mathrm{P} \left[ W_i = 1 \mid X_i \right]
