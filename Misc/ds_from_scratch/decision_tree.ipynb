**Q**: How is the entropy of a dataset $S$ consisting of class labels defined?

**A**: $$H(S) = -\sum_{i=1}^n p_i \log_2 p_i,$$

where $p_i$ is the proportion of examples in $S$ that belong to class $i$

---

**Q**: In a dataset consisting of class labels, when is entropy low and when is it high?

**A**: Entropy is low when most examples belong to one class, and high when the examples are distributed more evenly across classes.

---

**Q**: Write a function to compute entropy given class probabilities

**A**: See the function `entropy` below

---

```python
import math
from typing import List, Dict
import pandas as pd

def entropy(class_probs: List[float]) -> float:
    return -sum(p * math.log2(p) for p in class_probs if p > 0)
```

**Q**: Write a function to compute class probabilities for a dataset consisting of class labels

**A**: See the function `class_probs` below

---

```python
def class_probs(data: List[str]) -> List[float]:
    n = len(data)
    class_counts = Counter(data)
    return [count / n for count in class_counts.values()]
```

**Q**: Q: Write a function to compute the partition entropy of a labeled dataset split into subsets. Assume you have the functions `class_probs` and `entropy` at your disposal.

**A**: See the function `partition_entropy` below

---

```python
def partition_entropy(data_subsets: List[List[str]]) -> float:
    total_examples = sum(len(subset) for subset in data_subsets)
    class_probs_all = [class_probs(subset) for subset in data_subsets]
    subset_wts = [len(subset) / total_examples for subset in data_subsets]
    return sum(wt * entropy(cp) for wt, cp in zip(subset_wts, class_probs_all))
```

**Q**: How does the ID3 algorithm build a decision tree?

**A**: It recursively chooses the attribute with the lowest partition entropy, splits the data on that attribute, and repeats this process until all labels are the same or no attributes remain. 

---

**Q**: Write a function to compute the partition entropy of a label column in a pandas DataFrame after splitting the data by a categorical attribute. Assume you have the function partition_entropy at your disposal.

**A**: See the function `partition_entropy_by` function below

---

```python
def partition_entropy_by(
    data: pd.DataFrame,
    attribute: str,
    col_name: str
) -> float:
    partition_subsets = [
        subset[col_name].to_list()
        for _, subset in data.groupby(attribute)
    ]
    return partition_entropy(partition_subsets)
```
