**Q**: How is the entropy of a dataset $S$ consisting of class labels defined?

**A**: $$H(S) = -\sum_{i=1}^n p_i \log_2 p_i,$$

where $p_i$ is the proportion of examples in $S$ that belong to class $i$

---

**Q**: In a dataset consisting of class labels, when is entropy low and when is it high?

**A**: Entropy is low when most examples belong to one class, and high when the examples are distributed more evenly across classes.

---

**Q**: Write a function to compute entropy given class probabilities

**A**: See the function `entropy` below

---

```python
import math
from typing import List, Dict
import pandas as pd

def entropy(class_probs: List[float]) -> float:
    return -sum(p * math.log2(p) for p in class_probs if p > 0)
```

**Q**: Write a function to compute class probabilities for a dataset consisting of class labels

**A**: See the function `class_probs` below

---

```python
def class_probs(labels: List[str]) -> List[float]:
    n = len(labels)
    class_counts = Counter(labels)
    return [count / n for count in class_counts.values()]
```

**Q**: Write a function to compute the partition entropy of a dataset consisting of class labels split into subsets. Assume you have the functions `class_probs` and `entropy` at your disposal.

**A**: See the function `partition_entropy` below

---

```python
def partition_entropy(data_subsets: List[List[str]]) -> float:
    total_examples = sum(len(subset) for subset in data_subsets)
    class_probs_all = [class_probs(subset) for subset in data_subsets]
    subset_wts = [len(subset) / total_examples for subset in data_subsets]
    return sum(wt * entropy(cp) for wt, cp in zip(subset_wts, class_probs_all))
```

**Q**: How does the ID3 algorithm build a decision tree?

**A**: It recursively chooses the attribute with the lowest partition entropy, splits the data on that attribute, and repeats this process until all labels are the same or no attributes remain. 

---

**Q**: Write a function to compute the partition entropy of a label column in a pandas DataFrame after splitting the data by a categorical attribute. Assume you have the function `partition_entropy` at your disposal.

**A**: See the function `partition_entropy_by` function below

---

```python
def partition_entropy_by(
    data: pd.DataFrame,
    attribute: str,
    col_name: str
) -> float:
    partition_subsets = [
        subset[col_name].to_list()
        for _, subset in data.groupby(attribute)
    ]
    return partition_entropy(partition_subsets)
```

**Q**: How would you define `Leaf` and `Split` that represent a decision tree?

**A**: See below

---

```python
from typing import Any, NamedTuple, Union

class Leaf(NamedTuple):
    value: Any

class Split(NamedTuple):
    attribute: str
    subtrees: dict

DecisionTree = Union[Leaf, Split]
```

**Q:** In a decision tree, what do `attribute` and `subtrees` mean in `Split`?

**A:** `attribute` is the column we split on. `subtrees` maps each possible value to the next subtree.

---

**Q**: Write a function called `classify` that takes a `DecisionTree` as input and classifies an input `x`, where `x` is a `pd.Series` object

**A**: See the function `classify` below.

---

```python
def classify(tree: DecisionTree, x: pd.Series) -> Any:
    if isinstance(tree, Leaf):
        return tree.value

    attribute_val = x[tree.attribute]
    return classify(tree.subtrees[attribute_val], x)
```

**Q**: Write a function to build the decision tree classifier using the ID3 algorithm

**A**: See the function `build_tree_id3` below

---

```python
def build_tree_id3(
    data: pd.DataFrame,
    split_attributes: List[str],
    target_attribute: str
) -> DecisionTree:
    # check for leaf
    target_labels_subset = data[target_attribute]
    most_common_label = target_labels_subset.value_counts().idxmax()

    if target_labels_subset.nunique() == 1:
        return Leaf(most_common_label) 
    elif len(split_attributes) == 0:
        return Leaf(most_common_label)

    # find best attribute for splitting
    best_attribute = min(
        split_attributes, 
        key=lambda attr: partition_entropy_by(data, attr, target_attribute)
    )
    
    new_split_attributes = [attr for attr in split_attributes if attr != best_attribute]

    subtrees = {
        attr_val: build_tree_id3(
            subset,
            new_split_attributes,
            target_attribute
        )
        for attr_val, subset in data.groupby(best_attribute)
    }

    return Split(best_attribute, subtrees)
```

**Q**: What are the main characteristics of the ID3 algorithm?

**A**:

1. It builds a decision tree classifier.
2. It is designed for categorical features.
3. A split does not have to be binary.

---

**Q**: In the ID3 algorithm, why do we usually split on a column only once along a branch?

**A**: After splitting on a column, each resulting subset has the same value for that column, so that column cannot further separate the data along that branch.
