```python
from typing import NamedTuple, List
from collections import Counter
```

```python
def squared_distance(
    z: List[float],
    w: List[float]
) -> float:
    return sum((z_i - w_i)**2 for z_i, w_i in zip(z, w))
```

```python
def majority_vote(
    labels: List[str]
) -> str:
    vote_counts = Counter(labels)
    winner, winner_count = vote_counts.most_common(1)[0]

    n_winners = len([lb_count for lb_count in vote_counts.values() if lb_count == winner_count])

    if n_winners == 1:
        return winner

    return majority_vote(labels[:-1])
```

```python
class LabeledPoint(NamedTuple):
    point: List[float]
    label: str
```

```python
def knn_classify(
    data: List[LabeledPoint],
    input_point: List[float],
    k: int
) -> str:
    nn_lps = sorted(
        data, 
        key=lambda lp: squared_distance(lp.point, input_point)
    )
    knn_labels = [lp.label for lp in nn_lps[:k]]

    return majority_vote(knn_labels)
```
