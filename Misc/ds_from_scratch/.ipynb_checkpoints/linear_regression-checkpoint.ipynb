## Simple Linear Regression

**Q**: What is the general form of a simple regression model

**A**: $$y_i = \beta x_i + \alpha + \epsilon_i$$

---

#### Solving linear regression

Let $\left\{ (x_1, y_1), \ldots, (x_n, y_n) \right\}$ be the dataset. The least squares approximate solution is obtained by solving the equation:
$$
\left\| \mathbf{y} - \begin{bmatrix} \mathbf{1} & \mathbf{x} \end{bmatrix} \begin{bmatrix} \alpha \\ \beta \end{bmatrix} \right\|^2
$$

The linear least squares approximate solution is given by:
$$
\begin{align*}
\begin{bmatrix} \hat{\alpha} \\ \hat{\beta} \end{bmatrix} 
&= \left( \begin{bmatrix} \mathbf{1}^T \\[4pt] \mathbf{x}^T \end{bmatrix} \begin{bmatrix} \mathbf{1} & \mathbf{x} \end{bmatrix} \right)^{-1} \begin{bmatrix} \mathbf{1}^T \\[4pt] \mathbf{x}^T \end{bmatrix} \mathbf{y} \\[4pt]
&= \begin{bmatrix} n & n \bar{x} \\[4pt] n \bar{x} & \| \mathbf{x} \|^2 \end{bmatrix}^{-1} \begin{bmatrix} n \bar{y} \\[4pt] \mathbf{x}^T\mathbf{y} \end{bmatrix}
\end{align*}
$$

The inverse of the matrix
$$
\begin{bmatrix} n & n \bar{x} \\[4pt] n \bar{x} & \| \mathbf{x} \|^2 \end{bmatrix}
$$
is given by:
$$
\begin{align*}
\frac{1}{n \| \mathbf{x} \|^2 - n^2 \cdot \bar{x}^2} \cdot
\begin{bmatrix} \| \mathbf{x} \|^2 & -n \bar{x} \\[4pt] -n \bar{x} & n \end{bmatrix}
\end{align*}
$$

So we have:
$$
\begin{align*}
\begin{bmatrix} \hat{\alpha} \\ \hat{\beta} \end{bmatrix} 
&= \frac{1}{n \| \mathbf{x} \|^2 - n^2 \cdot \bar{x}^2} \cdot
\begin{bmatrix} \| \mathbf{x} \|^2 & -n \bar{x} \\[4pt] -n \bar{x} & n \end{bmatrix} \begin{bmatrix} n \bar{y} \\[4pt] \mathbf{x}^T\mathbf{y} \end{bmatrix} \\[4pt]
&= \frac{1}{n \| \mathbf{x} \|^2 - n^2 \cdot \bar{x}^2} \cdot 
\begin{bmatrix} n \bar{y} \cdot \| \mathbf{x} \|^2 - n \bar{x} \cdot \mathbf{x}^T\mathbf{y} \\[4pt] -n^2 \cdot \bar{x} \cdot \bar{y} + n \cdot \mathbf{x}^T\mathbf{y} \end{bmatrix}
\end{align*}
$$

So we get:
$$
\begin{align*}
\hat{\beta} &= \frac{-n^2 \cdot \bar{x} \cdot \bar{y} + n \cdot \mathbf{x}^T\mathbf{y}}{n \| \mathbf{x} \|^2 - n^2 \cdot \bar{x}^2} \\[4pt]
&= \frac{-n \cdot \bar{x} \cdot \bar{y} + \mathbf{x}^T\mathbf{y}}{\| \mathbf{x} \|^2 - n \cdot \bar{x}^2} \\[4pt]
&= \frac{\sum \left( x_i y_i - \bar{x} \bar{y} \right)}{\sum \left( x_i^2 - \bar{x}^2 \right)}
\end{align*}
$$

We have:
$$
\begin{align*}
\sum (x_i - \bar{x})(y_i - \bar{y}) &= \sum (x_i y_i - x_i \bar{y} - \bar{x} y_i + \bar{x} \bar{y}) \\[4pt]
&= \sum x_i y_i - \bar{y} \sum x_i - \bar{x} \sum y_i + \sum \bar{x} \bar{y} \\[4pt]
&= \sum x_i y_i - \bar{y}(n\bar{x}) - \bar{x}(n\bar{y}) + n\bar{x}\bar{y} \\[4pt]
&= \sum x_i y_i - n\bar{x}\bar{y} \\[4pt]
&= \sum (x_i y_i - \bar{x}\bar{y})
\end{align*}
$$

Similarly, for the denominator, we have:
$$
\begin{align*}
\sum (x_i - \bar{x})^2 &= \sum (x_i^2 - 2x_i\bar{x} + \bar{x}^2) \\[4pt]
&= \sum x_i^2 - 2\bar{x} \sum x_i + \sum \bar{x}^2 \\[4pt]
&= \sum x_i^2 - 2\bar{x}(n\bar{x}) + n\bar{x}^2 \\[4pt]
&= \sum x_i^2 - n\bar{x}^2 \\[4pt]
&= \sum (x_i^2 - \bar{x}^2)
\end{align*}
$$

So we get:
$$
\hat{\beta} = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}
$$

Dividing the numerator and denominator by $n-1$, we see that the numerator is an unbiased estimate of $\operatorname{cov}(x, y)$ and the denominator is an unbiased estimate of $\operatorname{var}(x)$. Thus, we can write:

$$
\hat{\beta} = \frac{\operatorname{cov}(x, y)}{\operatorname{var}(x)}
$$

From the normal equations, we have:
$$
\begin{bmatrix} n & n \bar{x} \\[4pt] n \bar{x} & \| \mathbf{x} \|^2 \end{bmatrix} \begin{bmatrix} \hat{\alpha} \\ \hat{\beta} \end{bmatrix} = \begin{bmatrix} n \bar{y} \\[4pt] \mathbf{x}^T\mathbf{y} \end{bmatrix}
$$

So we get:
$$
n\hat{\alpha} + n\bar{x}\hat{\beta} = n\bar{y},
$$

which implies:
$$
\hat{\alpha} = \bar{y} - \hat{\beta}\bar{x}
$$

**Q**: Given a simple linear regression model of the form:

$$y_i = \alpha + \beta x_i + \epsilon_i,$$

what are the values of $\alpha$ and $\beta$ that minimize the mean squared error over the dataset

**A**: 

$$
\hat{\beta} = \frac{\operatorname{cov}(x, y)}{\operatorname{var}(x)}
$$

and

$$
\hat{\alpha} = \bar{y} - \hat{\beta}\bar{x}
$$

**Q**: Intuitively, what does the $R$-squared metric represent?

**A**: How much better is our regression model compared to the constant estimator $\bar{y}$

---

**Q**: When does the $R$-squared metric become negative?

**A**: When our linear regression model performs worse than the constant estimator $\bar{y}$

---

**Q**: How is the $R$-squared metric defined?

**A**: $$R^2 = 1 - \frac{\text{SSE}}{\text{SST}},$$

where

$$\text{SSE} = \sum_{i=1}^n (y_i - \hat{y}_i)^2$$

and

$$\text{SST} = \sum_{i=1}^n (y_i - \bar{y})^2$$

---

**Q**: When is $R$-squared equal to 1 for a linear model?

**A**: When the sum of squared error over the dataset is equal to zero

---

**Q**: When is $R$-squared equal to 0 for a linear model?

**A**: When the model performs exactly the same as the constant estimator $\bar{y}$

---

```python
import numpy as  np

def generate_data():
    x = np.linspace(0, 1, 64)
    y = 2*x - 5 + np.random.randn(x.size)

    return x, y
```

```python
def sum_of_sqd_errors(y, y_hat):
    return np.sum(np.square(y - y_hat))
```

```python
def get_parameters(X, y):
    x_mean = np.mean(X)
    y_mean = np.mean(y)
    
    beta = np.sum((X - x_mean) * (y - y_mean)) / np.sum(np.square(X - x_mean))
    alpha = y_mean - beta * x_mean
    return alpha, beta
```

```python
def predict(x_hat, alpha, beta):
    return beta * x_hat + alpha
```

```python
import math

# generate data
X, y = generate_data()

# train-test split
idxs = np.arange(0, len(y))
np.random.shuffle(idxs)
split_idx = math.floor(0.8 * len(X)) # 80-20 split
train_idxs, test_idxs = idxs[:split_idx], idxs[split_idx:]
X_train, y_train = X[train_idxs], y[train_idxs]
X_test, y_test = X[test_idxs], y[test_idxs]

alpha, beta = get_parameters(X_train, y_train)
y_hat = predict(X_test, alpha, beta)

print("SSE: ", sum_of_sqd_errors(y_test, y_hat))
```

```python
def r_squared(y, y_hat):
    return 1 - np.sum(np.square(y - y_hat)) / np.sum(np.square(y - np.mean(y)))
```

```python
r_squared(y_test, y_hat)
```

## Multiple Linear Regression

**Q**: The general form of a multiple linear regression model is given by:
$$
y_i = \alpha + \beta_1 x_1^{(i)} + \cdots + \beta_k x_k^{(i)} + \epsilon_i
$$
What is the gradient of the mean squared error function?

**A**:
$$
\nabla \text{MSE} = -\frac{2}{N} \sum_{i=1}^N (y_i - \hat{y}_i) \cdot \mathbf{x}_i,
$$
where
$$
\mathbf{x}_i = \begin{bmatrix} 1 & x_1^{(i)} & \cdots & x_k^{(i)} \end{bmatrix},
$$
and
$$
\hat{y}_i = \alpha + \beta_1 x_1^{(i)} + \cdots + \beta_k x_k^{(i)}.
$$

**Explanation**: We have:
$$
\begin{align*}
\frac{\partial}{\partial \beta_j}\text{MSE}
&= \frac{\partial}{\partial \beta_j}\left(\frac{1}{N}\sum_{i=1}^N (y_i - \hat{y}_i)^2\right) \\[4pt]
&= -\frac{2}{N}\sum_{i=1}^N (y_i - \hat{y}_i)\frac{\partial \hat{y}_i}{\partial \beta_j} \\[4pt]
&= -\frac{2}{N}\sum_{i=1}^N (y_i - \hat{y}_i)x_j^{(i)}
\end{align*}
$$

---

**Q**: The general form of a multiple linear regression model is given by:
$$
y_i = \alpha + \beta_1 x_1^{(i)} + \cdots + \beta_k x_k^{(i)} + \epsilon_i
$$
Write a function using NumPy to compute the gradient of the mean squared error function

**A**: See the function `mse_grad` below

---

**Q**: The general form of a multiple linear regression model is given by:
$$
y_i = \alpha + \beta_1 x_1^{(i)} + \cdots + \beta_k x_k^{(i)} + \epsilon_i
$$
Write a function using NumPy to find the optimal parameters that minimize the MSE loss using mini-batch gradient descent. Assume you have the functions `mse_grad` and `mse_loss` at your disposal.

**A**: See the function `multi_least_squares_fit` below

---

```python
def mse_loss(X, y, params):
    preds = X @ params
    errors = y - preds
    return np.mean(np.square(errors))

def mse_grad(X, y, params):
    preds = X @ params
    errors = y - preds
    return np.mean(-2 * X * errors.reshape(-1, 1), axis=0)

def multi_least_squares_fit(
    X: np.ndarray, 
    y: np.ndarray, 
    n_iters: int=100, 
    batch_size: int=1,
    lr: float = 1e-2,
):
    assert X.shape[0] == len(y), "X and y should have the same number of rows"

    # note: X should be augmented with the constant 1
    n, p = X.shape
    n_batches = n // batch_size if n % batch_size == 0 else 1 + n // batch_size
    
    params = np.random.randn(p)

    for i in range(n_iters):
        total_avg_loss = 0
        for batch_start in range(0, n, batch_size):
            X_batch = X[batch_start: batch_start + batch_size]
            y_batch = y[batch_start: batch_start + batch_size]

            loss = mse_loss(X_batch, y_batch, params)
            grad = mse_grad(X_batch, y_batch, params)

            params -= lr*grad
            
            total_avg_loss += loss
            
        total_avg_loss = total_avg_loss / n_batches
        print(f"loss for iteration {i+1}: ", total_avg_loss)

    return params
```

**Q**: What is the intuitive idea behind the bootstrap

**A**: The idea is to create new datasets by choosing $n$ data points with replacement from the original dataset

---

**Q**: How can you estimate the standard error of a regression coefficient of a multiple regression model using bootstrap?

**A**: We create new datasets using the bootstrap and fit a regression model on each dataset. The standard deviation of a regression coefficient across the bootstrap estimates gives the standard error of that coefficient.

**Q**: The general form of a multiple linear regression model is given by:
$$
y_i = \alpha + \beta_1 x_1^{(i)} + \cdots + \beta_k x_k^{(i)} + \epsilon_i
$$
What is the gradient of the ridge penalty?

**A**:
$$
2\lambda \vec{\beta}
$$
where
$$
\vec{\beta} =
\begin{bmatrix}
0 \\
\beta_1 \\
\vdots \\
\beta_k
\end{bmatrix}
$$

**Explanation**: We have

$$
\frac{\partial}{\partial \beta_j}\left(\lambda \sum_{i=1}^k \beta_i^2\right) =
2\lambda \beta_j
$$
for $j = 1, \dots, k$

---

**Q**: The general form of a multiple linear regression model is given by:
$$
y_i = \alpha + \beta_1 x_1^{(i)} + \cdots + \beta_k x_k^{(i)} + \epsilon_i
$$
Write a function using NumPy to compute the gradient of the ridge penalty.

**A**: See the function `ridge_penalty_grad` below

---

```python
def ridge_penalty_grad(weight, params):
    grad = np.zeros(params.shape)
    grad[1:] = 2 * weight * params[1:]
    return grad
```

**Q**: The general form of a multiple linear regression model is given by:
$$
y_i = \alpha + \beta_1 x_1^{(i)} + \cdots + \beta_k x_k^{(i)} + \epsilon_i
$$
Write a function using NumPy to find the optimal parameters that minimize the ridge-regularized MSE loss using mini-batch gradient descent. Assume you have the functions `mse_grad`, `mse_loss_ridge`, and `ridge_penalty_grad` at your disposal.

**A**: See the function `multi_least_squares_fit_ridge` below

---

```python
def mse_loss_ridge(X, y, params, ridge_wt):
    preds = X @ params
    errors = y - preds
    return np.mean(np.square(errors)) + ridge_wt * np.sum(np.square(params[1:]))

def multi_least_squares_fit_ridge(
    X: np.ndarray, 
    y: np.ndarray, 
    n_iters: int=100, 
    batch_size: int=1,
    lr: float = 1e-2,
    ridge_wt: float = 0.2
):
    assert X.shape[0] == len(y), "X and y should have the same number of rows"

    # note: X should be augmented with the constant 1
    n, p = X.shape
    n_batches = n // batch_size if n % batch_size == 0 else 1 + n // batch_size
    
    params = np.random.randn(p)

    for i in range(n_iters):
        total_avg_loss = 0
        for batch_start in range(0, n, batch_size):
            X_batch = X[batch_start: batch_start + batch_size]
            y_batch = y[batch_start: batch_start + batch_size]

            loss = mse_loss_ridge(X_batch, y_batch, params, ridge_wt)
            grad = mse_grad(X_batch, y_batch, params)
            ridge_grad = ridge_penalty_grad(ridge_wt, params)

            params -= lr*(grad + ridge_grad)
            
            total_avg_loss += loss
            
        total_avg_loss = total_avg_loss / n_batches
        print(f"loss for iteration {i+1}: ", total_avg_loss)

    return params
```
