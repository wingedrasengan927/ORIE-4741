**Q**: Explain the behaviour of a well calibrated model

**A**: If the model predicts a patient has a 40% risk of death, then for a group of 100 similar patients who all get a predicted risk of 40%, about 40 of them should actually die.

**Q**: How is the brier score of a classifier defined?

**A**:
$$
\frac{1}{N} \sum_{i=1}^N \left( E_i - O_i \right)^2,
$$

where

- $N$ is the number of samples
- $E_i$ is the predicted estimate for sample $i$
- $O_i$ is the actual outcome for sample $i$

---

**Q**: How is the expected calibration error of a classifier defined?

**A**: First we sort the estimates and group them into $K$ bins with approximately an equal number of samples in each bin. Then, the ECE is defined as:
$$
\sum_{i=1}^K \mathrm{P}(i) \cdot \left| o_i - e_i \right|
$$

where

- $\mathrm{P}(i)$ is the fraction of all samples that fall into bin $i$
- $o_i$ is the fraction of positive instances in group $i$
- $e_i$ is the average of the probabilities for the instances in group $i$

---

**Q**: How is the maximum calibration error of a classifer defined?

**A**: First we sort the estimates and group them into $K$ bins with approximately an equal number of samples in each bin. Then, the ECE is defined as:
$$
\underset{i=1, \ldots, K}{\max} \left( \left| o_i - e_i \right| \right)
$$

where

- $o_i$ is the fraction of positive instances in group $i$
- $e_i$ is the average of the probabilities for the instances in group $i$

**Q**: What is the general idea of Platt scaling

**A**: The idea is to build a very simple, secondary logistic regression model that takes as input the predicted estimates of the primary model, and is trained to predict the true binary outcomes

---

**Q**: What is the general idea of Isotonic regression

**A**: The idea is to build a simple, secondary non-decreasing piece-wise constant model that takes as input the predicted estimates of the primary model, and is trained to predict the true binary outcomes

---

**Q**: What is the general idea behind quantile binning

**A**: In quantile binning, estimates are sorted and partitioned into $K$ bins of equal sizes. For every estimate within each bin, an estimate is calibrated to be equal to the fraction of positive samples in that bin.

---

**Q**: What is the general idea behind Bayesian Binning into Quantiles

**A**: BBQ is an ensemble method that combines multiple quantile binning models of different sizes by taking a weighted sum of their outputs
