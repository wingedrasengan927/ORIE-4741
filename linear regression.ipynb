- Total: $2mn -m + m + 2m - 1 = 2mn + 2m - 1$ flops

Q: Compute the gradient of the objective function $\| Ax - b \|^2$

A: We have:
$$
\begin{align*}
\| Ax - b \|^2 &= \left( Ax - b \right)^T \left( Ax - b \right) \\[4pt]
&= \left( x^TA^T - b^T \right) \left( Ax - b \right) \\[4pt]
&= x^TA^TAx - x^TA^Tb - b^TAx + b^Tb \\[4pt]
&= x^TA^TAx - 2x^TA^Tb + b^Tb \\[4pt]
\end{align*}
$$

Now the gradient can be computed directly as:
$$
\nabla f(x) = 2A^TAx - 2A^Tb
$$

Q: Show that the objective function defined by $f(x) = \| Ax - b \|^2$ is convex

A: Previously, we've computed the gradient of the function as:
$$
\nabla f(x) = 2A^TAx - 2A^Tb
$$
Now, the Hessian can be computed as:
$$
\nabla^2 f(x) = 2A^TA
$$
Thus the Hessian is positive semidefinite, which implies the function is convex

Q: Show that the objective function defined by $f(x) = \| Ax - b \|^2$ is L-smooth

A: It suffices to show that $\nabla f(x)$ is Lipschitz continuous. A function $g: \mathrm{R}^n \to \mathrm{R}^m$ is Lipschitz continuous if there exists a non-negative real number $L$ such that
$$
\| g(x_1) - g(x_2) \| \leq L \| x_1 - x_2 \|,
$$
for all $x_1, x_2 \in \mathrm{R}^n$

We have:
$$
\begin{align*}
\| \nabla f(x_1) - \nabla f(x_2) \| &= \| 2A^TAx_1 - 2A^Tb - 2A^TAx_2 + 2A^Tb \| \\[4pt]
&= \| 2A^TAx_1 - 2A^TAx_2 \| \\[4pt]
&= \| 2A^TA (x_1 - x_2) \| \\[4pt]
&\leq 2 \cdot \| A^T \| \cdot \| A \| \cdot \| x_1 - x_2 \| \\[4pt]
&= 2 \cdot \| A \| \cdot \| A \| \cdot \| x_1 - x_2 \| \\[4pt]
&= 2 \| A \|^2 \cdot \| x_1 - x_2 \| \\[4pt]
\end{align*}
$$

Q.E.D

Q: Consider the objective function $f(w) = \| y - Xw \|^2$. Show that $w$ minimises $f$ if and only if it satisfies the normal equations:
$$
X^TXw = X^Ty
$$

A: 

$\Rightarrow$

Suppose that $w$ minimises $f$. Since $f$ is convex, the gradient of $f$ at $w$ should be equal to zero. This implies:
$$
\begin{align*}
\nabla f(w) &= 0 \\[4pt]
\implies X^TXw - X^Ty &= 0 \\[4pt]
\implies X^TXw &= X^Ty
\end{align*}
$$
Q.E.D

$\Leftarrow$

Suppose that $w$ satisfies the normal equations. This implies
$$
\begin{align*}
X^TXw &= X^Ty \\[4pt]
\Rightarrow X^TXw - X^Ty &= 0 \\[4pt]
\Rightarrow 2X^TXw - 2X^Ty &= 0 \\[4pt]
\Rightarrow \nabla f(w) = 0 \\[4pt]
\end{align*}
$$

Consider a point $w^\prime$. Since $f$ is convex, we have:
$$
\begin{align*}
f(w^\prime) &\geq f(w) + \nabla f(w)^T (w^\prime - w) \\[4pt]
&= f(w)
\end{align*}
$$

Thus $w$ minimises $f$

Q.E.D

#### Convergence of Gradient Descent

We now show the objective function $f(x) = \| y - Xw \|^2$ decreases at each step, leading to convergence to the minimum value if the step size follows a specific condition

Previously, we've shown that $f(x)$ is L-smooth. This implies for any $x_1, x_2 \in \mathrm{R}^n$ the function $f(x)$ satisfies
$$
f(x_2) \leq f(x_1) + \nabla f(x_1)^T (x_2 - x_1) + \frac{L}{2} \| x_2 - x_1 \|^2,
$$
where $L = 2 \| X \|^2$

At step $w \in \mathrm{R}^n$, the update rule is given by:
$$
w^+ = w - \alpha \nabla f(w),
$$
where $\alpha > 0$ is the step size.

Using the L-smooth property of $f(x)$, we get:
$$
\begin{align*}
f(w^+) &\leq f(w) + \nabla f(w)^T (w^+ - w) + \frac{L}{2} \| w^+ - w \|^2 \\[4pt]
&= f(w) - \alpha \nabla f(w)^T \nabla f(w) + \frac{L}{2} \| \alpha \nabla f(w) \|^2 \\[4pt]
&= f(w) - \alpha \| \nabla f(w) \|^2 + \alpha^2\frac{L}{2} \|  \nabla f(w) \|^2 \\[4pt]
&= f(w) - \| \nabla f(w) \|^2 \left( \alpha - \alpha^2\frac{L}{2} \right) \\[4pt]
\end{align*}
$$
For $f$ to decrease, we need the term
$$
\left( \alpha - \alpha^2\frac{L}{2} \right)
$$
to be positive. Since $\alpha > 0$, this happens when:
$$
\alpha < \frac{2}{L}
$$

#### Solving least squares using QR factorisation

If $w$ minimises $\| y - Xw \|^2$, then it should satisfy the normal equations:
$$
X^TXw = X^Ty
$$

If $X^TX$ is invertible, then the solution is given by:
$$
w = \left( X^TX \right)^{-1} X^Ty
$$

However, forming $\left( X^TX \right)^{-1}$ is numerically unstable.

Let $X = QR$ be the QR factorisation of $X$. Since $X^TX$ is invertible, $X$ has linearly independent columns. This implies $R$ is invertible. Substituting $X=QR$ in the normal equations, we get:
$$
\begin{align*}
(QR)^TQRw &= (QR)^Ty \\[4pt]
\Rightarrow R^TQ^TQRw &= R^TQ^Ty \\[4pt]
\Rightarrow R^TRw &= R^TQ^Ty
\end{align*}
$$

Since $R$ is invertible and upper triangular, $R^T$ is also invertible. This implies that the above equation can be rewritten as:
$$
Rw = Q^Ty
$$

Since $R$ is upper triangular, $w$ can be solved using back substitution. We don't need to explicitly form the inverse of $R$
